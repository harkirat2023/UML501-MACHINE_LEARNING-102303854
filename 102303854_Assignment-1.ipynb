```python
# Q1(d)
x = np.array([1,2,3,4,5,1,2,1,1,1])
unique_x, counts_x = np.unique(x, return_counts=True)
most_frequent_x = unique_x[np.argmax(counts_x)]
indices_x = np.where(x == most_frequent_x)[0]
print(f"Array x: {x}")
print(f"Most frequent value: {most_frequent_x}")
print(f"Indices of mfv: {indices_x}")
print(f"Frequency: {np.max(counts_x)}\n")

y = np.array([1, 1, 1, 2, 3, 4, 2, 4, 3, 3])
unique_y, counts_y = np.unique(y, return_counts=True)
most_frequent_y = unique_y[np.argmax(counts_y)]
indices_y = np.where(y == most_frequent_y)[0]
print(f"\nArray y: {y}")
print(f"Most frequent value: {most_frequent_y}")
print(f"Indices ofmfv: {indices_y}")
print(f"Frequency: {np.max(counts_y)}")
```

```python
# Q1(e)
gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]')
print(f"Matrix gfg:\n{gfg}")
total_sum = np.sum(gfg)
print(f"\nSum of all elements: {total_sum}")
row_sum = np.sum(gfg, axis=1)
print(f"\nrow-wise Sum:\n{row_sum}")
col_sum = np.sum(gfg, axis=0)
print(f"\ncolumn-wise Sum:\n{col_sum}")
```

```python
# Q1(f)
n_array = np.array([[55, 25, 15],[30, 44, 2],[11, 45, 77]])
print(f"Matrix n_array:\n{n_array}")
diagonal_sum = np.trace(n_array)
print(f"\n Sum of diagonal elements: {diagonal_sum}")

eigenvalues, eigenvectors = np.linalg.eig(n_array)
print(f"\n Eigen values:\n{eigenvalues}")
print(f"\n Eigen vectors:\n{eigenvectors}")

matrix_inverse = np.linalg.inv(n_array)
print(f"\n Inverse of matrix:\n{matrix_inverse}")

determinant = np.linalg.det(n_array)
print(f"\n Determinant of matrix: {determinant}")
```

```python
# Q1(g)

p1 = np.array([[1, 2], [2, 3]])
q1 = np.array([[4, 5], [6, 7]])

print("FIRST PAIR:")
print(f"Matrix p1:\n{p1}")
print(f"Matrix q1:\n{q1}")

multiplication1 = np.dot(p1, q1)
print(f"\n Matrix multiplication p1 × q1:\n{multiplication1}")

covariance1 = np.cov(p1.flatten(), q1.flatten())
print(f"\n Covariance between p1 and q1:\n{covariance1}")

print("\n" + "*"*50)

p2 = np.array([[1, 2], [2, 3], [4, 5]])
q2 = np.array([[4, 5, 1], [6, 7, 2]])

print("\nSECOND PAIR:")
print(f"Matrix p2:\n{p2}")
print(f"Matrix q2:\n{q2}")

multiplication2 = np.dot(p2, q2)
print(f"\nMatrix multiplication p2 × q2:\n{multiplication2}")

covariance2 = np.cov(p2.flatten(), q2.flatten())
print(f"\nCovariance between p2 and q2:\n{covariance2}")
```

```python
# Q1(h)
import numpy as np
x = np.array([[2, 3, 4], [3, 2, 9]])
y = np.array([[1, 5, 0], [5, 10, 3]])

print(f"Matrix x:\n{x}")
print(f"Matrix y:\n{y}")

inner_product = np.inner(x, y)
print(f"\nInner product:\n{inner_product}")

outer_product = np.outer(x, y)
print(f"\nOuter product:\n{outer_product}")

x_flat = x.flatten()
y_flat = y.flatten()
cartesian_x, cartesian_y = np.meshgrid(x_flat, y_flat, indexing='ij')
cartesian_product = np.stack([cartesian_x.ravel(), cartesian_y.ravel()], axis=-1)
print(f"\nCartesian product (first 10 pairs):\n{cartesian_product[:10]}")
print(f"Total cartesian product pairs: {len(cartesian_product)}")
```

```python
# Q2(a)
import numpy as np
array = np.array([[1, -2, 3],[-4, 5, -6]])
abs_array = np.abs(array)
print(f"\nABSOLUTE VALUES:\n{abs_array}")

flattened = array.flatten()
print(f"\nPERCENTILES:")
print(f"Flattened array: {flattened}")

p25_flat = np.percentile(flattened, 25)
p50_flat = np.percentile(flattened, 50)
p75_flat = np.percentile(flattened, 75)
print(f"\nFlattened array - 25th percentile: {p25_flat}")
print(f"Flattened array - 50th percentile: {p50_flat}")
print(f"Flattened array - 75th percentile: {p75_flat}")

p25_col = np.percentile(array, 25, axis=0)
p50_col = np.percentile(array, 50, axis=0)
p75_col = np.percentile(array, 75, axis=0)
print(f"\nColumn-wise - 25th percentile: {p25_col}")
print(f"Column-wise - 50th percentile: {p50_col}")
print(f"Column-wise - 75th percentile: {p75_col}")

p25_row = np.percentile(array, 25, axis=1)
p50_row = np.percentile(array, 50, axis=1)
p75_row = np.percentile(array, 75, axis=1)
print(f"\nRow-wise - 25th percentile: {p25_row}")
print(f"Row-wise - 50th percentile: {p50_row}")
print(f"Row-wise - 75th percentile: {p75_row}")

print(f"\nSTATISTICS:")

mean_flat = np.mean(flattened)
median_flat = np.median(flattened)
std_flat = np.std(flattened)
print(f"\nFlattened array - Mean: {mean_flat}")
print(f"Flattened array - Median: {median_flat}")
print(f"Flattened array - Standard Deviation: {std_flat}")

mean_col = np.mean(array, axis=0)
median_col = np.median(array, axis=0)
std_col = np.std(array, axis=0)
print(f"\nColumn-wise - Mean: {mean_col}")
print(f"Column-wise - Median: {median_col}")
print(f"Column-wise - Standard Deviation: {std_col}")

mean_row = np.mean(array, axis=1)
median_row = np.median(array, axis=1)
std_row = np.std(array, axis=1)
print(f"\nRow-wise - Mean: {mean_row}")
print(f"Row-wise - Median: {median_row}")
print(f"Row-wise - Standard Deviation: {std_row}")
```

```python
# Q3(a)
import numpy as np
array = np.array([10, 52, 62, 16, 16, 54, 453])
print(f"Original array: {array}")

sorted_array = np.sort(array)
print(f"\nSorted array: {sorted_array}")

sorted_indices = np.argsort(array)
print(f"\nIndices of sorted array: {sorted_indices}")
print(f"Verification - array[sorted_indices]: {array[sorted_indices]}")

four_smallest_indices = np.argpartition(array, 4)[:4]
four_smallest = array[four_smallest_indices]
four_smallest_sorted = np.sort(four_smallest)
print(f"\n4 smallest elements: {four_smallest_sorted}")

five_largest_indices = np.argpartition(array, -5)[-5:]
five_largest = array[five_largest_indices]
five_largest_sorted = np.sort(five_largest)[::-1]
print(f"\n5 largest elements: {five_largest_sorted}")
```

```python
# Q3(b)
import numpy as np
array = np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0])
print(f"Original array: {array}")

integer_mask = array == array.astype(int)
integer_elements = array[integer_mask]
print(f"\nInteger elements only: {integer_elements}")

float_mask = array != array.astype(int)
float_elements = array[float_mask]
print(f"\nFloat elements only (non-integer): {float_elements}")

print(f"\nDetailed breakdown:")
print(f"Original array: {array}")
print(f"As integers: {array.astype(int)}")
print(f"Integer mask: {integer_mask}")
print(f"Float mask: {float_mask}")
```
